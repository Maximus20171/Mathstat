Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

Решение:

Среднее значение:
$$M(x)=\frac{a+b}{2}=\frac{200+800}{2}=500$$

Дисперсия:
$$D(x)=\frac{b-a}{12}=\frac{800-200}{12}=50$$

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

Решение:

$$D(x)=\frac{b-a}{12}$$ Зная что $a=0.5$ и $D(x)=0.2$ можно решить уровнение:
$$\frac{b-0.5}{12}=0.2$$

$$b=2.9$$

Теперь можно найти среднее:
$$M(x)=\frac{a+b}{2}=\frac{2.9+0.5}{2}=1.7$$

Непрерывная случайная величина X распределена нормально и задана плотностью распределения
f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-(x+2)**2) / 32).
Найдите:
а). M(X)
б). D(X)
в). std(X) (среднее квадратичное отклонение)

Решение:

Формула для нормального распредения:
$$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}}$$

Из формулы нормального распредения ясно что:

а)
$$M(x)=-2$$

б)
$$D(x)=16$$

в)
$$\sigma=4$$

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

Решение:
Будем сводить нормальное распределение к стандарному! и используя Z-таблица узнаем ответ.

##### а) больше 182 см

$$Z=\frac{182-174}{8}=1$$ т.е. $\sigma$. По Z-таблице значение $f(\sigma)=0.8413$. Но это значение показывает вероятность левой части распределния вероятности, а нам нужна левая, то находим ее $1-0.8413=0.1587$

##### б) больше 190 см

$$Z=\frac{190-174}{8}=2$$ т.е. 2$\sigma$. По Z-таблице значение $f(2\sigma)=0.9772$. Но это значение показывает вероятность левой части распределния вероятности, а нам нужна левая, то находим ее $1-0.9772=0.0228$

##### в) от 166 см до 190 см

$$Z_1=\frac{166-174}{8}=-1$$ т.е. -$\sigma$
$$Z_2=\frac{190-174}{8}=2$$ т.е. 2$\sigma$

по Z-таблице значение $f(-\sigma)=0.15866$.

по Z-таблице значение $f(2\sigma)=0.97725$.

от 166 до 190 = 0.97725-0.15866=0.81859

#####  г) от 166 см до 182 см

$$Z_1=\frac{166-174}{8}=-1$$ т.е. -$\sigma$
$$Z_2=\frac{182-174}{8}=1$$ т.е. $\sigma$

по Z-таблице значение $f(-\sigma)=0.15866$.

По Z-таблице значение $f(\sigma)=0.84134$

от 166 см до 182 см = 0.84134-0.15866=0.68268

Или по правилу 3-ех сигм 68% значений лежит в интервале от $-\sigma$ до $\sigma$


##### д) от 158 см до 190 см

$$Z_1=\frac{158-174}{8}=-2$$ т.е. -2$\sigma$
$$Z_2=\frac{190-174}{8}=2$$ т.е. 2$\sigma$

по Z-таблице значение $f(-2\sigma)=0.02275$.

по Z-таблице значение $f(2\sigma)=0.97725$.

от 158 см до 190 см = 0.97725-0.02275=0.9545

Или по правилу 3-ех сигм 95,4% значений лежит в интервале от $-2\sigma$ до $2\sigma$

##### е) не выше 150 см или не ниже 190 см

$$Z_1=\frac{150-174}{8}=-3$$ т.е. -3$\sigma$
$$Z_2=\frac{190-174}{8}=2$$ т.е. 2$\sigma$

по Z-таблице значение $f(-3\sigma)=0.00135$.

по Z-таблице значение $f(2\sigma)=0.97725$.

не ниже 190 см = 1-0.97725=0.02275

не выше 150 см или не ниже 190 см = 0.00135+0.02275=0.0241

##### ё) не выше 150 см или не ниже 198 см

$$Z_1=\frac{150-174}{8}=-3$$ т.е. -3$\sigma$
$$Z_2=\frac{198-174}{8}=3$$ т.е. 3$\sigma$

Проще решить используя правило 3х сигм. А точнее 99.72% значений лежит в интервале от -3$\sigma$ до 3$\sigma$.

не выше 150 см или не ниже 198 см = 100%-99,72%=0,28%=0,0028

##### ж) ниже 166 см.

$$Z=\frac{166-174}{8}=-1$$ т.е. -$\sigma$

по Z-таблице значение $f(-\sigma)=0.15866$.

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

Решение:

Найдем среденквадратичное распределение в этом распределении: $$\sigma=\sqrt{D(X)}=\sqrt{25}=5$$

Теперь найдем отклодение
$$Z=\frac{190-178}{5}=3\sigma$$
