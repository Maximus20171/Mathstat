Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.

Решение:

Формула нахождения довертельного интервала:

$$ \overline x ± (Z_{a/2}*\frac{\sigma}{\sqrt{n}})$$

подставялем:
$$80 ± (1.96 * \frac{16}{16})$$

т.е. доверительный интервал лежит в диапазоне $[78.04; 81.96]$

```python
# доверительный интервал
print(80-(1.96*1))
print(80+(1.96*1))
```

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

Решение:

Формула нахождения довертельного интервала ($\sigma$ не известна):

$$ \overline x ± (t_{a/2}*\frac{\sigma}{\sqrt n})$$

подставялем:
$$6.59 ± (2.262 * \frac{0,45}{\sqrt {10}})$$

т.е. значение Х лежит в доверительном интервале в диапазоне $[6.268; 6.912]$ с точностью 95%


```python
import numpy as np
a=[6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
```

```python
## средняя и дисперсия
print ('средняя M(a) =', np.sum(a)/len(a))
print ('стандарное отклонение =', np.std(a, ddof=1))
```

```python
# доверительный интервал
print(6.59-(2.262*np.std(a, ddof=1)/sqrt(10)))
print(6.59+(2.262*np.std(a, ddof=1)/sqrt(10)))
```

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.

Решение:

1)

$H_0: \mu = \mu_0$ 

$H_1: \mu > \mu_0$ 

2)

Критерий Z-тест т.к. известно дисперсия.

При α=0,05. $Z_0$ табличное = 1.65. Найдем Z выборки:

$$Z_н=\frac{\overline X - \mu_0}{\frac{\sigma}{\sqrt n}}$$

$$Z_н=\frac{17.5-17}{\frac{2}{10}}=\frac{0.5}{0.2}=2.5$$

т.к. $Z_н>Z_0$  то мы не можем прнять нулевую гипотизу. Т.е. с вероятностью 95% верна $H_1$
